```python
from gensim.models.keyedvectors import KeyedVectors

#TODO: remember to change the path to your needs.
word_vectors = KeyedVectors.load_word2vec_format("/content/drive/MyDrive/MN5002/GoogleNews-vectors-negative300.bin.gz", binary=True, limit=200000)
```

Now lets try out a few examples; what words are closest to cooking and potatoes?

```python
word_vectors.most_similar(positive=['cooking', 'potatoes'], topn=5)
```

We can also look at words that least match others in a list

```python
word_vectors.doesnt_match("potatoes milk cake computer".split())
```

Now, lets look at the famous "king + woman - man =" example

```python
word_vectors.most_similar(positive=['king','woman'], negative=['man'], topn=3)
```

How close are princess and queen?

```python
word_vectors.similarity('princess','queen')
```

What do the vocabulary frequencies look like?

```python
word_vectors.index_to_key[1000:1010]
```

What other things can we do with it?

```python
from umap import UMAP
data = word_vectors.vectors
print(data.shape)

reducer = UMAP(n_components=2, metric='cosine', n_neighbors = 15, min_dist=0.1, random_state = 12)
reduced_wv = reducer.fit_transform(data)
```
