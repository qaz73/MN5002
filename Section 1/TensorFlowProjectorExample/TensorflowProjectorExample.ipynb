This is just a short example of the power of the projection using Tensorflow Projector. Word vectors were created using [this example from Google Tensorflow](https://colab.research.google.com/github/tensorflow/docs/blob/master/site/en/tutorials/text/word2vec.ipynb#scrollTo=aj--8RFK6fgW) run in Colab. This works with word vectors based [the works of Shakespeare](https://storage.googleapis.com/download.tensorflow.org/data/shakespeare.txt)

The example created the two files
* **vectors.tsv** , word token vectors created using word2vec - roughly **6 MB** in size
* **metadata.tsv** , metadata describing the vectors (e.g., the text of the word associated with the vector) - roughly **30 kB** in size

Both of these files are provided in the Gitlab directory.

1. Download the two files to your own storage (local or Google drive).
1. In your browser, open [Tensorflow Projector](http://projector.tensorflow.org/).
1. On the upper left of the browser, click the **Load** icon
1. This will ask you to upload the two files **vectors.tsv** and **metadata.tsv** that you had downloaded
1. Now you can play with the various projections, including **UMAP**, **PCA** and **t-SNE**, to get an understanding about the associations between the words.
